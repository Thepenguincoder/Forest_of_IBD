```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Import the GO terms files downloaded from MGnify as well as the metadata
#I did not separate the GO terms into different types but just used all of them here
mgys2301 = pd.read_csv("MGYS00002301_GO_abundances_v4.1.tsv", sep="\t")
mgys4728 = pd.read_csv("MGYS00004728_GO_abundances_v4.1.tsv", sep="\t")
mgys3733 = pd.read_csv("MGYS00003733_GO_abundances_v4.1.tsv", sep="\t")
mgys3712 = pd.read_csv("MGYS00003712_GO_abundances_v4.1.tsv", sep="\t")
metadata = pd.read_csv("human_meta_disease.txt", sep="\t")
```

```python
#Merging the different studies into a single dataframe
from functools import reduce

#List of DataFrames to merge
dfs = [mgys3733, mgys3712, mgys2301, mgys4728]

#Merges based on GO, description, and category with an outer join
merged_df = reduce(lambda left, right: pd.merge(left, right, on=["GO", "description", "category"], how="outer"), dfs)
merged_df
```

```python
#remove all of the samples with less than 1000 reads total
columns_to_keep = [x for x in range(3,len(merged_df.columns)) if merged_df.iloc[:, x].sum() >= 100000]
columns_to_keep[:0] = [0,1,2]
filtered_merged_df = merged_df.iloc[:, columns_to_keep]
filtered_merged_df
```

```python
#Filters the dataframe based category, if you want the full dataset do not run this
#I did not use this for my final results, but its here if needed

rows_to_keep = [x for x in range(len(filtered_merged_df)) if filtered_merged_df.iloc[x, 2] =="cellular component"]
print(len(rows_to_keep))
filtered_merged_df = filtered_merged_df.iloc[rows_to_keep, :]
filtered_merged_df
```

```python
#Turn the table into one of relative abundances (its done this weird way to avoid int/float type warnings)

#First make a new dataframe with the GO, description, and category columns the same as before
norm_filtered_merged_df = pd.DataFrame()
norm_filtered_merged_df["GO"] = filtered_merged_df["GO"]
norm_filtered_merged_df["description"] = filtered_merged_df["description"]
norm_filtered_merged_df["category"] = filtered_merged_df["category"]

#Make a dictionary with the column names and the normalized column values for each column of samples
normalized_data = {
    column: filtered_merged_df[column] / filtered_merged_df[column].sum() 
    for column in filtered_merged_df.iloc[:,3:]
}

#Create a new DataFrame with the normalized columns
normalized_df = pd.DataFrame(normalized_data)

#Concatenate the dataframe with just GO, description, and category with the normalized columns
norm_filtered_merged_df = pd.concat([norm_filtered_merged_df, normalized_df], axis=1)
norm_filtered_merged_df
```

```python
#Remove the rows were none of the samples have more than 0.05% of the reads

rows_to_keep = [x for x in range(len(norm_filtered_merged_df)) if norm_filtered_merged_df.iloc[x][3:].max() >= 0.0005]
final_df = norm_filtered_merged_df.iloc[rows_to_keep, :]
final_df
```

```python
#Transpose the data so that the features are clumns and the samples are rows (makes training the forest easier)
filtered_final_df2 = final_df.T

#Makes the first row the column headers
filtered_final_df2.columns = filtered_final_df2.iloc[0] 
filtered_final_df2 = filtered_final_df2[1:] 

#Resets the index so that the first column become the run_ids
filtered_final_df2 = filtered_final_df2.reset_index()
filtered_final_df2.rename(columns={'index': 'run_id'}, inplace=True)

filtered_final_df2
```

```python
#Filters the metadata so only runs in the filtered data are included and resets the index
metadata_filtered = metadata.loc[metadata["run_id"].isin(filtered_final_df2["run_id"])]
metadata_filtered = metadata_filtered.reset_index(drop=True)

#Removes all columns that are not run_id or phenotype and removes any non healthy and non IBD samples
metadata_filtered = metadata_filtered[['run_id', 'phenotype']]
metadata_filtered = metadata_filtered.loc[~metadata_filtered["phenotype"].isin(["Parkinsons (gut)","Unknown (gut)"])]

metadata_filtered
```

```python
#merges the filtered data with the metadata on the run_id with an outer join
from functools import reduce
dfs = [filtered_final_df2, metadata_filtered]
final_labelled_df = reduce(lambda left, right: pd.merge(left, right, on=["run_id"], how="inner"), dfs)

#changes healthy and diseased into numerical representations
final_labelled_df.replace('Healthy (gut)', 0, inplace=True)
final_labelled_df.replace('IBD (gut)', 1, inplace=True)
final_labelled_df
```

```python
#Removes any columns with NaN values
#This could be handeled differently but because the gridsearch did not work with NaN values I decided to handle it this way
final_labelled_df = final_labelled_df.dropna(axis=1, how='any')
final_labelled_df
```

```python
#saves the final dataframe as a CSV file
final_labelled_df.to_csv("GO_terms_processed.csv", index=False)
```
