```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Import the phylum taxonomy files downloaded from MGnify as well as the metadata
mgys2301 = pd.read_csv("MGYS00002301_phylum_SSU_v4.1.tsv", sep="\t")
mgys4728 = pd.read_csv("MGYS00004728_phylum_SSU_v4.1.tsv", sep="\t")
mgys3733 = pd.read_csv("MGYS00003733_phylum_SSU_v4.1.tsv", sep="\t")
mgys3712 = pd.read_csv("MGYS00003712_phylum_SSU_v4.1.tsv", sep="\t")
metadata = pd.read_csv("human_meta_disease.txt", sep="\t")
```

```python
#Merging the different studies into a single dataframe
from functools import reduce

#List of DataFrames to merge
dfs = [mgys3733, mgys3712, mgys2301, mgys4728]

#Merges based on phylum and kingdom with an outer join
merged_df = reduce(lambda left, right: pd.merge(left, right, on=["phylum", "kingdom"], how="outer"), dfs)

merged_df
```

```python
#remove all of the samples with less than 1000 reads total
columns_to_keep = [x for x in range(2,len(merged_df.columns)) if merged_df.iloc[:, x].sum() >= 1000]
columns_to_keep[:0] = [0,1]
semi_filtered_df = merged_df.iloc[:, columns_to_keep]

#Remove all the rows that are not Bacteria
filtered_df = semi_filtered_df[semi_filtered_df["kingdom"] == "Bacteria"]
filtered_df = filtered_df.drop(filtered_df.columns[0], axis=1)
filtered_df
```

```python
#Turn the table into one of relative abundances (its done this weird way to avoid int/float type warnings)

#First make a new dataframe with the phylum column the same as before
filtered_norm_df = pd.DataFrame()
filtered_norm_df["phylum"] = filtered_df["phylum"]

#Make a dictionary with the column names and the normalized column values for each column of samples
normalized_data = {
    column: filtered_df[column] / filtered_df[column].sum() 
    for column in filtered_df.iloc[:,1:]
}

#Create a new DataFrame with the normalized columns
normalized_df = pd.DataFrame(normalized_data)

#Concatenate the dataframe with just the phylums with the normalized columns
filtered_norm_df = pd.concat([filtered_norm_df, normalized_df], axis=1)
filtered_norm_df
```

```python
#Remove the phylums were none of the samples have more than 0.1% of the reads
rows_to_keep = [x for x in range(len(filtered_norm_df)) if filtered_norm_df.iloc[x][1:].max() >= 0.001]
filtered_final_df = filtered_norm_df.iloc[rows_to_keep, :]
filtered_final_df
```

```python
#Transpose the data so that the features are clumns and the samples are rows (makes training the forest easier)
filtered_final_df2 = filtered_final_df.T

#Makes the first row the column headers
filtered_final_df2.columns = filtered_final_df2.iloc[0] 
filtered_final_df2 = filtered_final_df2[1:] 

#Resets the index so that the first column become the run_ids
filtered_final_df2 = filtered_final_df2.reset_index()
filtered_final_df2.rename(columns={'index': 'run_id'}, inplace=True)

filtered_final_df2
```

```python
#Filters the metadata so only runs in the filtered data are included and resets the index
metadata_filtered = metadata.loc[metadata["run_id"].isin(filtered_final_df2["run_id"])]
metadata_filtered = metadata_filtered.reset_index(drop=True)

#Removes all columns that are not run_id or phenotype and removes any non healthy and non IBD samples
metadata_filtered = metadata_filtered[['run_id', 'phenotype']]
metadata_filtered = metadata_filtered.loc[~metadata_filtered["phenotype"].isin(["Parkinsons (gut)","Unknown (gut)"])]

metadata_filtered
```

```python
#merges the filtered data with the metadata on the run_id with an outer join
from functools import reduce
dfs = [filtered_final_df2, metadata_filtered]
final_labelled_df = reduce(lambda left, right: pd.merge(left, right, on=["run_id"], how="inner"), dfs)

#changes healthy and diseased into numerical representations
final_labelled_df.replace('Healthy (gut)', 0, inplace=True)
final_labelled_df.replace('IBD (gut)', 1, inplace=True)

final_labelled_df
```

```python
#Removes columns with NaN values (In this case there were only columns that had more than 50% NaN so I decided to remove them)
final_labelled_df = final_labelled_df.dropna(axis=1, how='any')
final_labelled_df
```

```python
#saves the final dataframe as a CSV file
final_labelled_df.to_csv("Relative_abundances_processed.csv", index=False)

```
